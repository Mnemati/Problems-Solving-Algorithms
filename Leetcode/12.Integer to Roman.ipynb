```python
class Solution(object):

        

    def intToRoman(self, num):
        """
        :type num: int
        :rtype: str
        """
        map1 = {0:'', 1:'I', 5:'V', 10:'X', 50: 'L', 100:'C', 500: 'D', 1000:'M'}
        map2 = {4:'IV', 9:'IX', 40:'XL', 90:'XC', 400:'CD', 900:'CM'}
        out = []
        power = 1
        while num > 0:
            digit = num % 10
            s = ''
            if power == 1:
                number = digit*power
                if number in map2:
                    out.append(map2[number])
                else:
                    r = number % 5
                    q = (number // 5)*5
                    if q!= 0:
                        s+=(map1[q]) 
                    if r != 0:
                        s+=(int(r/power) * map1[1])
                out.append(s)

            elif power == 10:
                number = digit*power
                if number in map2:
                    out.append(map2[number])
                else:
                    r = number % 50
                    q = (number // 50)*50
                    if q !=0:
                        s+=(map1[q]) 
                    if r!=0:
                        s+=(int(r/power) * map1[10])
                out.append(s)
            elif power == 100:
                number = digit*power
                if number in map2:
                    out.append(map2[number])
                else:
                    r = number % 500
                    q = (number // 500)*500
                    if q !=0:
                        s+=(map1[q]) 
                    if r!=0:
                        s+=(int(r/power) * map1[100])   
                out.append(s)  
                pass
            elif power > 100:
                number = digit*power
                if number in map2:
                    out.append(map2[number])
                else:
                    r = number % 1000
                    q = (number // 1000)*1000
                    if q !=0:
                        s+=(int(q/power)*map1[1000]) 
                    if r!=0:
                        s+=(int(r/power) * map1[1000])
                out.append(s)
            print(out)
            power *= 10
            num = num // 10
        
        out.reverse()
        out = ''.join(out)



        return out

    

            

        
```
